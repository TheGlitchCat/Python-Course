```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import pylab
from scipy import misc
from skimage.feature import local_binary_pattern
from skimage.feature import canny
from skimage import io
from skimage import color
import imageio
%matplotlib inline
```

```python
bins = range(0, 250)
```

```python
img = imageio.imread('img1.png')

R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]
img_gray = R * 299. / 1000 + G * 587. / 1000 + B * 114. / 1000

img_gray = img_gray.astype('int32')
plt.imshow(img_gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
img2 = imageio.imread('img2.png')

R2 = img2[:, :, 0]
G2 = img2[:, :, 1]
B2 = img2[:, :, 2]
img2_gray = R2 * 299. / 1000 + G2 * 587. / 1000 + B2 * 114. / 1000

img2_gray = img2_gray.astype('int32')
plt.imshow(img2_gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
img3 = imageio.imread('img3.png')

R3 = img3[:, :, 0]
G3 = img3[:, :, 1]
B3 = img3[:, :, 2]
img3_gray = R3 * 299. / 1000 + G3 * 587. / 1000 + B3 * 114. / 1000

img3_gray = img3_gray.astype('int32')
plt.imshow(img3_gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
plt.hist(img_gray.ravel(), bins, color="grey")
plt.title("Histograma RGB - Chihuahua")
plt.show()
```

```python
plt.hist(img2_gray.ravel(), bins, color="grey")
plt.title("Histograma 2 RGB - Muffin 1")
plt.show()
```

```python
plt.hist(img3_gray.ravel(), bins, color="grey")
plt.title("Histograma RGB - Muffin 2")
plt.show()
```

```python
img_red = R.astype('int32')
plt.imshow(img_red, cmap = plt.get_cmap('Reds'))
plt.show()

img2_red = R2.astype('int32')
plt.imshow(img2_red, cmap = plt.get_cmap('Reds'))
plt.show()

img3_red = R3.astype('int32')
plt.imshow(img3_red, cmap = plt.get_cmap('Reds'))
plt.show()
```

```python
#bins = np.linspace(-10, 10, 100, 1000)

plt.hist(img_red.ravel(), bins, color="red", alpha=0.5, label='img1')  # arguments are passed to np.histogram
plt.title("Histograma del Componente Rojo - Chihuahua")
plt.show()

plt.hist(img2_red.ravel(), bins, color="#7f0000", alpha=0.5, label='img2')  # arguments are passed to np.histogram
plt.title("Histograma del Componente Rojo - Muffin 1")
plt.show()

plt.hist(img3_red.ravel(), bins, color="#af0000", alpha=0.5, label='img2')  # arguments are passed to np.histogram
plt.title("Histograma del Componente Rojo - Muffin 2")
#plt.legend(loc='upper left')
plt.show()
```

```python
img_green = G.astype('int32')
plt.imshow(img_green, cmap = plt.get_cmap('Greens'))
plt.show()

img2_green = G2.astype('int32')
plt.imshow(img2_green, cmap = plt.get_cmap('Greens'))
plt.show()

img3_green = G3.astype('int32')
plt.imshow(img3_green, cmap = plt.get_cmap('Greens'))
#plt.show()
```

```python
plt.hist(img_green.ravel(), bins, color="green")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Verde - Chihuahua")
plt.show()

plt.hist(img2_green.ravel(), bins, color="green")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Verde - Muffin 1")
plt.show()

plt.hist(img3_green.ravel(), bins, color="green")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Verde - Muffin 2")
plt.show()
```

```python
img_blue = B.astype('int32')
plt.imshow(img_blue, cmap = plt.get_cmap('Blues'))
plt.show()

img2_blue = B2.astype('int32')
plt.imshow(img2_blue, cmap = plt.get_cmap('Blues'))
plt.show()

img3_blue = B3.astype('int32')
plt.imshow(img3_blue, cmap = plt.get_cmap('Blues'))
plt.show()
```

```python
plt.hist(img_blue.ravel(), bins, color="blue")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Azul - Chihuahua")
plt.show()

plt.hist(img2_blue.ravel(), bins, color="blue")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Azul - Muffin 1")
plt.show()

plt.hist(img3_blue.ravel(), bins, color="blue")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Azul - Muffin 2")
plt.show()
```

```python
radius = 3
n_points = 8 *radius

METHOD = 'default'
lbp = local_binary_pattern(img_gray, n_points, radius, METHOD)
lbp = lbp.astype('int32')
plt.imshow(lbp,cmap = plt.get_cmap('gray'))
plt.show()

lbp2 = local_binary_pattern(img2_gray, n_points, radius, METHOD)
lbp2 = lbp2.astype('int32')
plt.imshow(lbp2,cmap = plt.get_cmap('gray'))
plt.show()

lbp3 = local_binary_pattern(img3_gray, n_points, radius, METHOD)
lbp3 = lbp3.astype('int32')
plt.imshow(lbp3,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
lbp = (lbp - lbp.min())/(lbp.max() - lbp.min()) * 255.0


plt.hist(lbp.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de textura - Chihuahua")
plt.show()

lbp2 = (lbp2 - lbp2.min())/(lbp2.max() - lbp2.min()) * 255.0


plt.hist(lbp2.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de textura - Muffin 1")
plt.show()


lbp3 = (lbp3 - lbp3.min())/(lbp3.max() - lbp3.min()) * 255.0


plt.hist(lbp3.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de textura - Muffin 2")
plt.show()

```

```python
dx = scipy.ndimage.sobel(img_gray, 0)  # derivada horizontal
dy = scipy.ndimage.sobel(img_gray, 1)  # derivada vertical
mag = np.hypot(dx, dy)  # magnitud
mag *= 255.0 / np.max(mag)  # normalización
plt.imshow(mag, cmap = plt.get_cmap('gray'))
plt.show()

d2x = scipy.ndimage.sobel(img2_gray, 0)  # derivada horizontal
d2y = scipy.ndimage.sobel(img2_gray, 1)  # derivada vertical
mag2 = np.hypot(d2x, d2y)  # magnitud
mag2 *= 255.0 / np.max(mag2)  # normalización
plt.imshow(mag2, cmap = plt.get_cmap('gray'))
plt.show()

d3x = scipy.ndimage.sobel(img3_gray, 0)  # derivada horizontal
d3y = scipy.ndimage.sobel(img3_gray, 1)  # derivada vertical
mag3 = np.hypot(d3x, d3y)  # magnitud
mag3 *= 255.0 / np.max(mag3)  # normalización
plt.imshow(mag3, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
plt.hist(mag.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de bordes - Chihuahua")
plt.show()

plt.hist(mag2.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de bordes - Muffin 1")
plt.show()

plt.hist(mag3.ravel(), bins, color="black")  # arguments are passed to np.histogram
plt.title("Histograma de bordes - Muffin 2")
plt.show()
```
