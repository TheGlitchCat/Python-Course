## Activation Function

- ReLu: Allows all positive values to pass through without changing but assigns all negative values to 0

```python
import tensorflow as tf 
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fashion = keras.datasets.fashion_mnist
(train_imgs, train_labels), (test_imgs, test_labels) = fashion.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot' ]
```

```python
train_imgs.shape
```

```python
plt.figure()
plt.imshow(train_imgs[25])
plt.grid()
```

```python
train_imgs = train_imgs / 255.5
test_imgs = test_imgs / 255.5
```

```python
plt.figure(figsize = (10,10))

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_imgs[i], cmap='binary')
    plt.xlabel(class_names[train_labels[i]])
```

```python
# input_shape(28,28) = size of our imgs
# dense = relation of the pixels 

model = keras.Sequential([keras.layers.Flatten(input_shape = (28, 28)), 
                          keras.layers.Dense(128, activation = tf.nn.relu), 
                          keras.layers.Dense(10, activation = tf.nn.softmax)])
```

```python
model.compile(optimizer = "adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
# epoch = number of iterations
model.fit(train_imgs, train_labels, epochs = 5)
```

```python
test_loss, test_acc = model.evaluate( test_imgs, test_labels )
print(f'Accuracy: {test_acc}')
```

```python
predictions = model.predict(test_imgs)
```

```python
plt.figure(figsize = (10,10))

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    
    predicted_label = np.argmax(predictions[i])
    true_label = test_labels[i]
    if predicted_label == true_label:
        plt.imshow(test_imgs[i], cmap='binary')
        color='blue'
    else:
        plt.imshow(test_imgs[i], cmap=plt.cm.Accent)
        color='red'
    
    plt.xlabel('{} ({})'.format(class_names[predicted_label], class_names[true_label]), color=color)
```
